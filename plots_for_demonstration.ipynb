# Some figures used in the demonstration figure

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from my_code.plotting import graph_theme, ibm
```

```python
means = [-1, 0, 2, ]
stds = [0.5, 1, 1.5, ]

n = 10000

data = pd.DataFrame()

data = pd.concat(
    [
        pd.DataFrame(
            np.random.normal(mean, std, n), 
            columns=[f'value_{mean}']
        )
        for mean, std in zip(means, stds)
    ], 
    axis=1
)
data = pd.melt(data)
```

```python
with graph_theme():
    plt.close()

    fig, ax = plt.subplots(1,1,figsize=(5,3))

    sns.kdeplot(
        data=data, 
        x='value', 
        hue='variable', 
        alpha=0.65,
        legend=False,
        ax=ax,
        fill=True,
        palette=ibm[:3],
    )

    sns.kdeplot(
        data=data, 
        x='value', 
        legend=False,
        ax=ax,
        color="xkcd:salmon",
        linewidth=2,
        linestyle='--',
    )


    ax.set_xlabel('')
    ax.set_ylabel('')

    ax.set_yticklabels([])
    ax.set_xticklabels([])


    fig.subplots_adjust(left=0.01, right=0.99, top=0.99, bottom=0.01)

    fig.savefig('results/normal_mixture.pdf', transparent=True)
    fig.savefig('results/normal_mixture.svg', transparent=True)


    plt.show()
```
